```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Infection count - COVID-19_Radiography_Dataset
CovMeta = pd.read_excel('COVID-19_Radiography_Dataset/COVID.metadata.xlsx')
NormalMeta = pd.read_excel('COVID-19_Radiography_Dataset/normal.metadata.xlsx')
LungOpacMeta = pd.read_excel('COVID-19_Radiography_Dataset/Lung_Opacity.metadata.xlsx')
VirPneumMeta = pd.read_excel('COVID-19_Radiography_Dataset/Viral Pneumonia.metadata.xlsx')
all_metadata = [NormalMeta, CovMeta, VirPneumMeta, LungOpacMeta]
labels = ['Healthy', 'COVID', 'Viral Pneumonia', 'Other Infection']
row_counts = [df.shape[0] for df in all_metadata]
plt.figure(figsize=(10, 6))
smallset_color = (0.6, 0.3, 0.2)
bars = plt.bar(labels, row_counts, color = smallset_color, edgecolor = 'black')
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height/2),
                 xytext=(0, 0),
                 textcoords='offset points',
                 ha='center', va='center',
                 color='black',
                 fontsize=14)
plt.title('Infection type count')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```

```python
# Infection count - COVID-QU-Ex Dataset
COVID_test = os.listdir("COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Test\COVID-19\images")
COVID_train = os.listdir("COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Train\COVID-19\images")
COVID_val = os.listdir("COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Val\COVID-19\images")
COVID_count = len(COVID_test) + len(COVID_train) + len(COVID_val)
Non_COVID_test = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Test\Non-COVID\images")
Non_COVID_train = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Train\Non-COVID\images")
Non_COVID_val = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Val\Non-COVID\images")
Non_COVID_count = len(Non_COVID_test) + len(Non_COVID_train) + len(Non_COVID_val)
Normal_test = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Test\Normal\images")
Normal_train = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Train\Normal\images")
Normal_val = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Val\Normal\images")
Normal_count = len(Normal_test) + len(Normal_train) + len(Normal_val)
print(Normal_count, COVID_count, Non_COVID_count)
row_counts = [COVID_count, Non_COVID_count, Normal_count]
labels = ['Covid', 'Non_Covid','Healthy']
plt.figure(figsize=(10, 6))
bigset_color = (0.4, 0.7, 0.8)
bars = plt.bar(labels, row_counts, color = bigset_color, edgecolor = 'black')
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, 6000),
                 xytext=(0, 0),
                 textcoords='offset points',
                 ha='center', va='bottom',
                 color='black',
                 fontsize=14)
plt.title('Infection type count')
plt.xlabel('Infection')
plt.ylabel('Count')
plt.show()
```

```python
# Infection count - Both Datasets
CovMeta = pd.read_excel('COVID-19_Radiography_Dataset/COVID.metadata.xlsx')
NormalMeta = pd.read_excel('COVID-19_Radiography_Dataset/normal.metadata.xlsx')
LungOpacMeta = pd.read_excel('COVID-19_Radiography_Dataset/Lung_Opacity.metadata.xlsx')
VirPneumMeta = pd.read_excel('COVID-19_Radiography_Dataset/Viral Pneumonia.metadata.xlsx')
Non_COVID_Infection = pd.concat([LungOpacMeta, VirPneumMeta], ignore_index=True)
all_metadata = [NormalMeta, CovMeta, Non_COVID_Infection]

row_counts_smallset = [df.shape[0] for df in all_metadata]

COVID_test = os.listdir("COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Test\COVID-19\images")
COVID_train = os.listdir("COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Train\COVID-19\images")
COVID_val = os.listdir("COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Val\COVID-19\images")
COVID_count = len(COVID_test) + len(COVID_train) + len(COVID_val)
Non_COVID_test = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Test\Non-COVID\images")
Non_COVID_train = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Train\Non-COVID\images")
Non_COVID_val = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Val\Non-COVID\images")
Non_COVID_count = len(Non_COVID_test) + len(Non_COVID_train) + len(Non_COVID_val)
Normal_test = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Test\Normal\images")
Normal_train = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Train\Normal\images")
Normal_val = os.listdir(r"COVID-QU-Ex Dataset\Lung Segmentation Data\Lung Segmentation Data\Val\Normal\images")
Normal_count = len(Normal_test) + len(Normal_train) + len(Normal_val)
print(Normal_count, COVID_count, Non_COVID_count)

row_counts_bigset = [Normal_count, COVID_count, Non_COVID_count]

plt.figure(figsize=(10, 6))

bar_width = 0.35
labels = ['Healthy', 'COVID', 'Non-COVID Infection']

x_smallset = np.arange(len(labels))
smallset_color = (0.6, 0.3, 0.2)
x_bigset = x_smallset + bar_width
bigset_color = (0.4, 0.7, 0.8)

bars_smallset = plt.bar(x_smallset, row_counts_smallset, width=bar_width, label='COVID-19_Radiography_Dataset', color = smallset_color, edgecolor = 'black')
bars_bigset = plt.bar(x_bigset, row_counts_bigset, width=bar_width, label='COVID-QU-Ex Dataset', color = bigset_color, edgecolor = 'black')

for bars in [bars_smallset, bars_bigset]:
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height/2),
                 xytext=(0, 0),
                 textcoords='offset points',
                 ha='center', va='center',
                 color='black',
                 fontsize=14)
    
plt.xticks(x_smallset + bar_width / 2, labels)
plt.title('Dataset size per condition')
plt.xlabel('Infection')
plt.ylabel(' ')
plt.legend()
plt.show()
```

```python
# Images origin for the Normal subdataset of COVID-19_Radiography_Dataset
NormalMeta = pd.read_excel('COVID-19_Radiography_Dataset/normal.metadata.xlsx')
origin_counts = NormalMeta.groupby('URL').size()
plt.figure(figsize=(10, 6))
bars = plt.bar(origin_counts.index, origin_counts.values, edgecolor = 'black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}', ha='center', va='bottom')

plt.title('Origin of images from the Healthy dataset')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xticks(ticks=range(len(origin_counts)), labels=["RSNA", "Guangzhou"], rotation=0)
plt.show()
```

```python
# Images origin for the Covid subdataset of COVID-19_Radiography_Dataset
CovMeta = pd.read_excel('COVID-19_Radiography_Dataset/COVID.metadata.xlsx')
origin_counts = CovMeta.groupby('URL').size()
plt.figure(figsize=(10, 6))
bars = plt.bar(origin_counts.index, origin_counts.values, edgecolor = 'black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}', ha='center', va='bottom')

plt.title('Origin of images from the COVID dataset')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xticks(ticks=range(len(origin_counts)), labels=["BIMCV", "Eurorad", "CXnet paper", "Chestray GitHub", "Image Repo", "SIRM"], rotation=0)
plt.show()
```
